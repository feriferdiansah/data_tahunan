```python
# Import library
from matplotlib import pyplot as plt

# Plot sederhana awal
x = [10, 30, 50]
y0 = [20, 30, 40]
y = [25, 15, 30]

plt.plot(x, y, color="green", label="y (green)")
plt.plot(x, y0, label="y0 (blue)")

plt.title("Contoh Plot")
plt.xlabel("Nilai x")
plt.ylabel("Nilai y")
plt.legend()
plt.grid(True)
plt.show()

# Data rekapitulasi
data_rekap = [
    {"tahun": 1995, "penjualan": 300, "operasional": 150, "pajak": 50, "pendapatan_lain": 135},
    {"tahun": 1997, "penjualan": 320, "operasional": 120, "pajak": 70, "pendapatan_lain": 105},
    {"tahun": 1991, "penjualan": 200, "operasional": 75, "pajak": 35, "pendapatan_lain": 85},
    {"tahun": 1992, "penjualan": 320, "operasional": 170, "pajak": 80, "pendapatan_lain": 235},
    {"tahun": 2001, "penjualan": 300, "operasional": 150, "pajak": 50, "pendapatan_lain": 135},
    {"tahun": 2003, "penjualan": 500, "operasional": 350, "pajak": 100, "pendapatan_lain": 55},
    {"tahun": 2005, "penjualan": 750, "operasional": 450, "pajak": 200, "pendapatan_lain": 155}
]

# Memisahkan data untuk grafik
data_graph = {
    "tahun": [d["tahun"] for d in data_rekap],
    "penjualan": [d["penjualan"] for d in data_rekap],
    "pajak": [d["pajak"] for d in data_rekap],
    "operasional": [d["operasional"] for d in data_rekap],
    "pendapatan_lain": [d["pendapatan_lain"] for d in data_rekap],
    # Menghitung laba = penjualan + pendapatan_lain - pajak - operasional
    "laba": [
        d["penjualan"] + d["pendapatan_lain"] - d["pajak"] - d["operasional"]
        for d in data_rekap
    ]
}

# Plot 1: Grafik Laba dan Penjualan
plt.plot(data_graph["tahun"], data_graph["laba"], color="blue", marker="o", label="Laba")
plt.plot(data_graph["tahun"], data_graph["penjualan"], color="green", marker="o", label="Penjualan")

plt.title("Grafik Laba dan Penjualan")
plt.xlabel("Tahun")
plt.ylabel("Nilai (dalam satuan)")
plt.legend()
plt.grid(True)
plt.show()

# Plot 2: Grafik Laba, Penjualan, Pajak, Pendapatan Lain, dan Operasional
plt.plot(data_graph["tahun"], data_graph["laba"], color="blue", marker="o", label="Laba")
plt.plot(data_graph["tahun"], data_graph["penjualan"], color="green", marker="o", label="Penjualan")
plt.plot(data_graph["tahun"], data_graph["pajak"], color="red", marker="x", label="Pajak")
plt.plot(data_graph["tahun"], data_graph["operasional"], color="orange", marker="x", label="Operasional")
plt.plot(data_graph["tahun"], data_graph["pendapatan_lain"], color="purple", marker="x", label="Pendapatan Lain")

plt.title("Grafik Laba, Penjualan, Pajak, Pendapatan Lain, dan Operasional")
plt.xlabel("Tahun")
plt.ylabel("Nilai (dalam satuan)")
plt.legend()
plt.grid(True)
plt.show()
```
